## ADDING ROWS TO A DATAFRAME

The idea is to make the collection of new rows into a new dataframe. Then you just "glue" that new dataframe onto the end of your original dataframe.


```python
# to create a new dataframe, start with a dictionary
new_rows_dict = {
    'TIME' : [2000,2000,2000],
    'GEO' : ['USA', 'Canada', 'Uganda'],
    'Value' : [5, 4.18, 5.15]    
}

# turns the dictionary into a dataframe
new_rows_df = pd.DataFrame(new_rows_dict)

new_rows_df
```

```python
# Append the new_rows_df to the end of edu
edu_new_rows = edu.append(new_rows_df, ignore_index=True, sort=False)

edu_new_rows.tail()
```

```python
# Worksheet q4

# set up a distionary of rows you want added
ws4_dict = {
    'TIME' : [1502],
    'GEO' : ['Country X'],
    'Value' : [56]
}

# turn that into a dataframe
ws4_df = pd.DataFrame(ws4_dict)

# append that df to the old dataframe
ws4_new_df = edu_new_rows.append(ws4_df, ignore_index=True, sort=False)

ws4_new_df.tail()
```

# Filtering Dataframes

Boolean filtering

```python
# return the rows with Value bigger than 4
edu[ edu['Value'] > 4 ]
```

```python
# return the rows with Value between 4 and 6
edu[ (edu['Value'] > 4) & (edu['Value'] < 6) ]
```

```python
# return rows for years 2002 and 2003
edu[ (edu['TIME'] == 2002) | (edu['TIME'] == 2003)]
```

```python
# filter some of the columns
edu[ edu['TIME'] == 2002 ][["GEO", "Value"]]
```

```python
# return on the rows with missing Value data
# edu[ edu['Value'].isnull() ]

# return the rows with NO missing data
edu[ ( edu['Value'].notnull() ) | ( edu['GEO'].notnull() ) ]
```

# Aggregating Data

how can we sum, average, or otherwise aggregate data that shares a similar property?

**Standard aggregating functions from numpy**

count() -- this function counts how many values there are

sum() -- adds the values of interest

mean() -- finds the mean of the aggregate values

median()

min() -- finds the minimum value

max() -- finds the maximum value

prod() -- the multiplies all the values

std() -- returns the standard deviation of all the values

var() -- returns the variance of all the value


```python
group_mean = edu[['GEO', 'Value']].groupby('GEO').mean() # finds the mean Value within each country
print group_mean
```

```python
group_max = edu[['TIME', 'Value']].groupby('TIME').sum()
print group_max
```

# Pivot Tables

A pivot table is similar to these tables that we got using .groupby() except that instead of aggregating over one variable, we aggregate over two variables.  

```python
ptable = pd.pivot_table(
    edu,
    values = 'Value',
    index = 'TIME',
    columns = 'GEO',
    aggfunc = np.sum
)

ptable
```

# Plotting Data

The basic format to plot data is:

dataframe.plot(x = "Column1", y= "Column2", kind = GraphType, title = 'Title' )

```python
# here's scatter plot
edu.plot(x = 'TIME', y = 'Value', kind = 'scatter', title = 'Education')
```

```python
# here's a line plot

edu_austria = edu[ edu['GEO'] == 'Austria' ]

edu_austria.plot(x = 'TIME', y = 'Value', kind = 'line', title = 'edu')
```

```python
# bar graphs
# a bar grpah of the total sum value within each geo region

edu_country_value_aggregate = edu[['GEO','Value']].groupby('GEO').max()

edu_country_value_aggregate.plot(kind = 'bar')

```

```python
# bar graphs
# a bar grpah of the total sum value within each geo region

edu_country_value_aggregate = edu[['GEO','Value']].groupby('GEO').sum()

edu_country_value_aggregate.plot(kind = 'barh')
```
